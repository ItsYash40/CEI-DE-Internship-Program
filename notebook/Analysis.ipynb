2: Missing Value Handling

```python
# Check null values

df.isnull().sum()
```

```python
# Percentage of missing values

(df.isnull().sum() / len(df)) * 100
```

```python
# Fill numeric columns with median

numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)


# Fill categorical columns with mode

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)


print("Missing Values Handled")
```

```python
# Verify missing values

df.isnull().sum()
```

3: Remove Duplicate Records

```python
# Check duplicates

print("Duplicate rows:")

df.duplicated().sum()
```

```python
# Remove duplicates

df.drop_duplicates(inplace=True)

print("Duplicates Removed")

df.shape
```

6: Feature Engineering

```python
# ==========================================
# Feature Engineering
# ==========================================

import pandas as pd


# Clean price columns properly

df['initial_price'] = (
    df['initial_price']
    .astype(str)
    .str.replace('[^0-9.]', '', regex=True)
)

df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace('[^0-9.]', '', regex=True)
)


# Convert to numeric

df['initial_price'] = pd.to_numeric(
    df['initial_price'],
    errors='coerce'
)

df['final_price'] = pd.to_numeric(
    df['final_price'],
    errors='coerce'
)


# Fill missing prices

df['initial_price'].fillna(
    df['initial_price'].median(),
    inplace=True
)

df['final_price'].fillna(
    df['final_price'].median(),
    inplace=True
)


# Create price difference column

df['price_difference'] = (
    df['initial_price']
    -
    df['final_price']
)


# Convert rating columns

df['rating'] = pd.to_numeric(
    df['rating'],
    errors='coerce'
)

df['ratings_count'] = pd.to_numeric(
    df['ratings_count'],
    errors='coerce'
)


# Fill missing values

df['rating'].fillna(0, inplace=True)
df['ratings_count'].fillna(0, inplace=True)


# Popularity Score

df['popularity_score'] = (
    df['rating']
    *
    df['ratings_count']
)


# Create quantity because dataset doesn't have it

df['quantity'] = 1


# total_amount = price * quantity

df['total_amount'] = (
    df['final_price']
    *
    df['quantity']
)


# Show result

df[
    [
        'title',
        'initial_price',
        'final_price',
        'price_difference',
        'rating',
        'ratings_count',
        'popularity_score',
        'total_amount'
    ]
].head()
```

7: Univariate Analysis

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['final_price'],
    bins=30,
    kde=True
)

plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['rating'],
    bins=20,
    kde=True
)

plt.title("Rating Distribution")

plt.show()
```

8: Bivariate Analysis

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df['final_price'],
    y=df['rating']
)

plt.title(
    "Price vs Rating Relationship"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df['price_difference'],
    y=df['popularity_score']
)

plt.title(
    "Discount vs Popularity"
)

plt.show()
```

9: Category Level Analysis

```python
# Average price by category

category_price = (
    df.groupby('category')
    ['final_price']
    .mean()
    .sort_values(
        ascending=False
    )
)


category_price
```

```python
plt.figure(figsize=(10,5))

category_price.head(10).plot(
    kind='bar'
)

plt.title(
    "Top Categories by Average Price"
)

plt.ylabel(
    "Average Price"
)

plt.show()
```

10: Boxplot Analysis

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=df['final_price']
)

plt.title(
    "Price Outlier Detection"
)

plt.show()
```
